# Notebook to calculate grid of equally-scaped points on tilted / non-orthogonal axes

### Import packages 

```python
# Import required packages: numpy for mathematical operations and data arrays, pandas for data tables, and matplotlib.pyplot for data visualisation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define x and y coordinates of points A, B and C
Ax = 293011.18
Ay = 4582418.90
Bx = 293242.5
By = 4582483.72
Cx = 292955.65
Cy = 4582621.82
```

```python
# Create numpy arrays for each location using its x,y coordinates 
A = np.array([Ax, Ay])
B = np.array([Bx, By])
C = np.array([Cx, Cy])
```

```python
# Define horizontal and vertical axis (directions) as vectors 
h = B - A
v = C - A
```

```python
# Calculate the norm of h and v
h_norm = np.sqrt(np.sum(np.power(h,2)))
v_norm = np.sqrt(np.sum(np.power(v,2)))
```

```python
# Calculate the cosine of angles alpha and gamma 
α_cos = h[0]/h_norm
α_sin = h[1]/h_norm

γ_cos = v[0]/v_norm
γ_sin = v[1]/v_norm
```

```python
# Define point spacing 
s = 6 
```

```python
dx = s * α_cos
dy = s * α_sin
```

```python
dh = 6 
dv = 6

hs = dh*68
vs = dv*42

l_hs = -dh*10
l_vs = -dv*3

origin = A
```

```python
# Create grid of points
xs, ys = [], []
for i in np.arange(l_hs, hs, dh):
    for j in np.arange(l_vs, vs, dv):
        x = origin[0] + (i * α_cos) + (j * γ_cos)
        y = origin[1] + (i * α_sin) + (j * γ_sin)
        xs.append(x)
        ys.append(y)
```

### Plot figure

```python
plt.figure(figsize=(10,10)) # define size of figure
plt.scatter(xs,ys,color='yellow') # scatter plot of xs and ys 
plt.scatter(A[0],A[1],color='red',marker='x') # scatter plot of location A
plt.scatter(B[0],B[1],color='green',marker='x') # scatter plot of location B
plt.scatter(C[0],C[1],color='blue',marker='x') # scatter plots of location C 
plt.show() # show plot
```

### Export to CSV 

```python
df = pd.DataFrame() # create empty pandas dataframe  
df['X'] = xs # fill column X of pandas dataframe with xs
df['Y'] = ys # fill column Y of pandas dataframe with ys

path = 'name.csv' # define path name for CSV   
df.to_csv(path, index=False) # export dataframe to CSV
```
